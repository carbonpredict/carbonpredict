```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype
```

```python
DATA_PATH = '../mnt_emission_data' # Need to have the data as CSV files in this path
```

```python
content = sorted(filter(lambda x: x.endswith(".csv"), os.listdir(DATA_PATH)))

df = pd.concat((pd.read_csv(f'{DATA_PATH}/{f}') for f in content))
df
```

```python
df_co2 = df[~df["co2_total"].isna()]
```

```python
df_co2.to_csv("../testdata/all_co2.csv", index=False)
```

```python
df_noweight = df_co2[~df_co2["weight"].isna()]
```

```python
df_noweight.to_csv("../testdata/all_co2_no_weight.csv", index=False)
```

```python
df_co2_small = df_co2[:100000]
```

```python
df_co2_small.to_csv("../testdata/co2_100000.csv", index=False)
```
